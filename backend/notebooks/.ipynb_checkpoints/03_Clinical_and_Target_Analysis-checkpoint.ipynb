```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load cleaned dataset
df = pd.read_csv("Bacteria_dataset_cleaned.csv")

# Quick overview
print(df.shape)
df.head()
```

```python
print(f"Rows    : {df.shape[0]}")
print(f"Columns : {df.shape[1]}")

df.info()

missing_summary = pd.DataFrame({
    "Missing Count": df.isna().sum(),
    "Missing Percentage": (df.isna().mean() * 100).round(2)})
missing_summary

df.describe(include="all").T
```

```python
clinical_features = [
    "Age",
    "Gender",
    "Diabetes",
    "Hypertension",
    "Hospital_before",
    "Infection_Freq"
]
```

```python
plt.figure(figsize=(8,5))
sns.histplot(
    data=df,
    x="Age",
    bins=20,
    kde=True)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

plt.figure(figsize=(8,2))
sns.boxplot(
    x=df["Age"])
plt.title("Age Boxplot")
plt.show()
```

```python
def analyze_binary_feature(df, column):
    counts = df[column].value_counts(dropna=False)
    percentages = (
        df[column]
        .value_counts(dropna=False, normalize=True)
        .mul(100)
        .round(2)
    )
    summary = pd.DataFrame({
        "Count": counts,
        "Percentage": percentages
    })
    print(summary)
    plt.figure(figsize=(5,4))
    sns.countplot(
        data=df,
        x=column
    )
    plt.title(f"{column} Distribution")
    plt.xlabel(column)
    plt.ylabel("Count")
    plt.show()
```

```python
analyze_binary_feature(df, "Gender")
analyze_binary_feature(df, "Diabetes")
analyze_binary_feature(df, "Hypertension")
analyze_binary_feature(df, "Hospital_before")
```

```python
infection_summary = pd.DataFrame({
    "Count": df["Infection_Freq"].value_counts(dropna=False),
    "Percentage": (
        df["Infection_Freq"]
        .value_counts(dropna=False, normalize=True)
        *100
    ).round(2)})
infection_summary
```
